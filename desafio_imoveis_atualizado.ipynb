## 2. üßÆ C√°lculo do Valor por Metro Quadrado (m¬≤)

````python

# Remover im√≥veis com √°rea zero para evitar erro de divis√£o
df = df[df["Area_m2"] > 0]

# Criar nova coluna com valor por m¬≤
df["Valor_m2"] = df["Aluguel_Mensal"] / df["Area_m2"]
# .head() mostra as primeiras 5 linhas do DataFrame
df[["Aluguel_Mensal", "Area_m2", "Valor_m2"]].head()
````

## 3. üèôÔ∏è Selecionando Bairros com Alta Demanda de Loca√ß√£o

````python

# Selecionando apenas im√≥veis dispon√≠veis
df_disponiveis = df[df["Disponivel"] == "Sim"]

# Contando im√≥veis por cidade e bairro
demanda_bairros = (
# .groupby() agrupa os dados por uma ou mais colunas
    df_disponiveis.groupby(["Cidade", "Bairro"])
# .size() conta quantas ocorr√™ncias existem em cada grupo
    .size()
    .reset_index(name="Qtd_Imoveis_Disponiveis")
)

# Selecionando bairros com mais de 50 im√≥veis dispon√≠veis
bairros_alta_demanda = demanda_bairros[demanda_bairros["Qtd_Imoveis_Disponiveis"] > 50]
# .sort_values() ordena o DataFrame pela coluna escolhida
bairros_alta_demanda.sort_values(by="Qtd_Imoveis_Disponiveis", ascending=False)
````

## 4. üîó Cruzando os Dados com Informa√ß√µes Completas

````python

# Juntando os dados de alta demanda com os im√≥veis dispon√≠veis
# .merge() combina dois DataFrames com base nas colunas indicadas
df = df_disponiveis.merge(bairros_alta_demanda, on=["Cidade", "Bairro"], how="inner")

# Visualizar dados completos
# .head() mostra as primeiras 5 linhas do DataFrame
df.head()
````

## 5. üè∑Ô∏è Classificando o Valor por m¬≤

````python

# # A fun√ß√£o √© feita geralmente nessa estrutura, mas vou deixar
# # coment√°rios mais detalhados para o entendimento de voc√™s.
# def classificar_valor_m2(valor):
#     if valor <= 20:
#         return "Baixo"
#     elif valor <= 40:
#         return "M√©dio"
#     else:
#         return "Alto"

# # .apply() aplica uma fun√ß√£o personalizada a cada valor da coluna
# df["Classificacao_Valor_m2"] = df["Valor_m2"].apply(classificar_valor_m2)
# # .value_counts() conta a frequ√™ncia de cada valor √∫nico na coluna
# df["Classificacao_Valor_m2"].value_counts()
````

````python
# ‚úÖ Etapa 1: Criar uma fun√ß√£o para classificar os valores do m¬≤ (Valor_m2)

# Esta fun√ß√£o recebe como entrada o valor do metro quadrado (valor)
# e retorna uma categoria textual: "Baixo", "M√©dio" ou "Alto",
# de acordo com faixas de pre√ßo estabelecidas.

def classificar_valor_m2(valor):
    # Se o valor for menor ou igual a 20 reais por m¬≤, classificamos como "Baixo"
    if valor <= 20:
        return "Baixo"
    
    # Se o valor estiver entre 20 e 40 (inclusive), classificamos como "M√©dio"
    elif valor <= 40:
        return "M√©dio"
    
    # Se o valor for maior que 40, classificamos como "Alto"
    else:
        return "Alto"


# ‚úÖ Etapa 2: Aplicar a fun√ß√£o a cada linha da coluna "Valor_m2"

# A fun√ß√£o .apply() do pandas permite aplicar uma fun√ß√£o personalizada
# a cada valor individual de uma coluna.

# Aqui, estamos criando uma nova coluna chamada "Classificacao_Valor_m2"
# onde cada linha ser√° classificada de acordo com o valor do m¬≤ correspondente.
df["Classificacao_Valor_m2"] = df["Valor_m2"].apply(classificar_valor_m2)


# ‚úÖ Etapa 3: Contar quantos im√≥veis existem em cada faixa de classifica√ß√£o

# A fun√ß√£o .value_counts() retorna a quantidade de vezes que cada valor
# aparece na coluna especificada. Aqui, queremos saber:
# - Quantos im√≥veis est√£o na faixa "Baixo"
# - Quantos est√£o em "M√©dio"
# - E quantos est√£o em "Alto"

df["Classificacao_Valor_m2"].value_counts()
````

## 6. üìä Gr√°fico: Top 10 Bairros com Maior Valor M√©dio do m¬≤

````python

media_valor_m2 = (
# .groupby() agrupa os dados por uma ou mais colunas
    df.groupby(["Cidade", "Bairro"])["Valor_m2"]
    .mean()
    .reset_index(name="Media_Valor_m2")
# .sort_values() ordena o DataFrame pela coluna escolhida
    .sort_values(by="Media_Valor_m2", ascending=False)
)

top_bairros = media_valor_m2.head(10)

plt.figure(figsize=(12,6))
# plt.barh() cria um gr√°fico de barras horizontal
plt.barh(top_bairros["Bairro"] + " - " + top_bairros["Cidade"], top_bairros["Media_Valor_m2"])
plt.xlabel("Valor m√©dio por m¬≤ (R$)")
plt.title("Top 10 bairros com maior valor m√©dio por m¬≤")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
````

## 7. üåê Gr√°fico Interativo: √Årea vs Aluguel

````python

# üìà Gr√°fico de linha interativo: valor m√©dio do m¬≤ por bairro

# Passo 1: calcular a m√©dia do valor_m2 agrupado por cidade e bairro
media_valor_m2 = (
    df.groupby(["Cidade", "Bairro"])["Valor_m2"]
    .mean()
    .reset_index(name="Media_Valor_m2")
    .sort_values(by="Media_Valor_m2", ascending=False)
)

# Passo 2: criar uma coluna combinando Cidade e Bairro para facilitar a visualiza√ß√£o
media_valor_m2["Local"] = media_valor_m2["Cidade"] + " - " + media_valor_m2["Bairro"]

# Passo 3: gr√°fico de linha interativo

fig = px.line(
    media_valor_m2,
    x="Local",
    y="Media_Valor_m2",
    markers=True,
    title="üìà Valor M√©dio do m¬≤ por Bairro (Interativo)",
)

# Melhorar layout do gr√°fico
fig.update_layout(xaxis_tickangle=-45, xaxis_title="Bairro", yaxis_title="Valor m√©dio do m¬≤ (R$)", height=500)

fig.show()

````

‚úÖ O que esse gr√°fico mostra?
Um ponto para cada bairro (combinado com a cidade).

O eixo Y mostra o valor m√©dio por m¬≤.

Ao passar o mouse, voc√™ v√™ os detalhes de cada ponto.

Os pontos est√£o ligados por linhas (com markers=True) para facilitar o acompanhamento visual.

Ao executar a celula do gr√°fico e der certo, ele vai abrir o navegador e mostrar o gr√°fico.
